```python
import pandas as pd
from numpy import *
import seaborn as sns
from matplotlib.pyplot import *
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)
n=15
x=np.linspace(0,10,n) + np.random.randn(n)/5
y=np.sin(x)+x/6 + np.random.randn(n)/10

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

# Question No 01 
# Reshape DataSet

```python
x_test = x_test.flatten()
x_test=x_test.reshape((len(x_test), 1))

y_test = y_test.flatten()
y_test=y_test.reshape((len(y_test), 1))


x_train = x_train.flatten()
x_train=x_train.reshape((len(x_train), 1))

y_train = y_train.flatten()
y_train=y_train.reshape((len(y_train), 1))
```

# Plot Training Dataset

```python
plt.subplots(figsize=(13, 5)) 
plt.scatter(x_train,y_train,color="blue")
plt.scatter(x_test,y_test,color="red")
plt.title('Traing Dataset',fontsize=15)
plt.xlabel('x_train',fontsize=15)
plt.ylabel('y_train',fontsize=15)
plt.legend(['Training dataset','Testing dataset'])
plt.show()
```

# Call Polynomial Object And Fit x_train

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
degree=[1,3,6,9]
for i in degree:
    poly_reg=PolynomialFeatures(degree=i)
    x_poly=poly_reg.fit_transform(x_train)
    
    reg=LinearRegression()
    reg.fit(x_poly,y_train)
    
    
    plt.subplots(figsize=(13, 5)) 
    plt.scatter(x_train,y_train,color="blue")
    plt.scatter(x_test,y_test,color="red")
#     plot(x_train,reg.predict(poly_reg.fit_transform(x_train)),color="green")
    plt.scatter(x_train,reg.predict(poly_reg.fit_transform(x_train)),color="green")
    plt.title('Polynomail',fontsize=15)
    plt.xlabel('x_train',fontsize=15)
    plt.ylabel('y_train',fontsize=15)
    plt.legend(['Training dataset','Testing dataset','prediction'])
    plt.show()
```

# Plot Polynomial Points With Testing Dataset

```python
reg.score(x_poly,y_train)
```

# Question no 02

```python
x_test = x_test.flatten()
x_test=x_test.reshape((len(x_test), 1))
y_test = y_test.flatten()
y_test=y_test.reshape((len(y_test), 1))

output_traindata=[]
output_testdata=[]

# output_traindata=tuple()
# output_testdata=

for i in range(0,10):
    poly_reg=PolynomialFeatures(degree=i)
    
    x_poly=poly_reg.fit_transform(x_train)
    testx_poly=poly_reg.fit_transform(x_test)
    
    regtrain=LinearRegression()
    regtrain.fit(x_poly,y_train)
    
    regtest=LinearRegression()
    regtest.fit(testx_poly,y_test)
    
    traindata_score=regtrain.score(x_poly,y_train)
    traindata_cofficient=regtrain.coef_
    
    testdata_score=regtest.score(testx_poly,y_test)
    testdata_cofficient=regtest.coef_
    
    output_traindata.append({traindata_score,traindata_cofficient[0][0]})
    output_testdata.append({testdata_score,testdata_cofficient[0][0]})

test_final=list(tuple(d) for d in output_testdata)
test_final
```

```python
train_final=list(tuple(d) for d in output_traindata)
train_final
```

```python
print("Shape of TEST: {} and Shape of TRAIN {}".format(shape(test_final),shape(train_final)))
```
